```python
import csv

# train1_analysis1 = pd.read_csv('./metrics/analysis1/train_case1_metrics.csv').drop(columns=['Unnamed: 0'])
# train2_analysis1 = pd.read_csv('./metrics/analysis1/train_case2_metrics.csv').drop(columns=['Unnamed: 0'])
test1_analysis1 = pd.read_csv('./metrics/analysis1/test_case1_metrics.csv').drop(columns=['Unnamed: 0'])
test2_analysis1 = pd.read_csv('./metrics/analysis1/test_case2_metrics.csv').drop(columns=['Unnamed: 0'])

# train1_analysis2 = pd.read_csv('./metrics/analysis2/train_case1_metrics.csv').drop(columns=['Unnamed: 0'])
# train2_analysis2 = pd.read_csv('./metrics/analysis2/train_case2_metrics.csv').drop(columns=['Unnamed: 0'])
test1_analysis2 = pd.read_csv('./metrics/analysis2/test_case1_metrics.csv').drop(columns=['Unnamed: 0'])
test2_analysis2 = pd.read_csv('./metrics/analysis2/test_case2_metrics.csv').drop(columns=['Unnamed: 0'])


# train1_analysis3 = pd.read_csv('./metrics/analysis3/train_case1_metrics.csv').drop(columns=['Unnamed: 0'])
# train2_analysis3 = pd.read_csv('./metrics/analysis3/train_case2_metrics.csv').drop(columns=['Unnamed: 0'])
test1_analysis3 = pd.read_csv('./metrics/analysis3/test_case1_metrics.csv').drop(columns=['Unnamed: 0'])
test2_analysis3 = pd.read_csv('./metrics/analysis3/test_case2_metrics.csv').drop(columns=['Unnamed: 0'])
```

```python

valores_modelo_a_analysis1 = list(test1_analysis1.values[0])
valores_modelo_b_analysis1 = list(test2_analysis1.values[0])

valores_modelo_a_analysis2 = list(test1_analysis2.values[0])
valores_modelo_b_analysis2 = list(test2_analysis2.values[0])

valores_modelo_a_analysis3 = list(test1_analysis3.values[0])
valores_modelo_b_analysis3 = list(test2_analysis3.values[0])
```

```python
sns.set(font_scale=2)

modelos = ['Caso 1 - Sem Preenchimento', 'Caso 2 - Sem Preenchimento', 'Caso 1 - Com SMOTE', 'Caso 2 - Com SMOTE', 'Caso 1 - Com dados Reais', 'Caso 2 - Com dados Reais']
metricas = ['Recall', 'Precisão', 'Especificidade', 'Acurácia', 'F1-Score']
valores_modelos = [
    valores_modelo_a_analysis1,
    valores_modelo_b_analysis1,
    valores_modelo_a_analysis2,
    valores_modelo_b_analysis2,
    valores_modelo_a_analysis3,
    valores_modelo_b_analysis3
]

width = 0.12
posicao_metricas = np.arange(len(metricas))


fig, ax = plt.subplots(figsize=(20,10))
rects = []
for i, modelo in enumerate(modelos):
    posicao_modelo = posicao_metricas + (i - 2.5) * width
    rects.append(ax.bar(posicao_modelo, valores_modelos[i], width, label=modelo))

ax.set_ylabel('Valor')
ax.set_title('Comparação de Métricas de Performance')
ax.set_ylim(0,1.2)
ax.set_xticks(posicao_metricas)
ax.set_xticklabels(metricas)
ax.legend(loc='lower right')

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom',
                    rotation=90)

for rect in rects:
    autolabel(rect)

plt.tight_layout()
plt.show()
```
